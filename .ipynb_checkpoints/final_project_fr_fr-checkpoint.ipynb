## Notebook Purpose

The goal of our project is to map the flood events ranging from year 2000 to 2018. More specifically, the goal is to focus on the flood events in Bangladesh and their impacts in the past 20 years. Bangladesh has a tropical monsoon-type climate, with hot and rainy summers and dry winters. It has very low-lying land, and is exposed to increased rainfall and cyclones that are likely going to intensify due to climate change. Bangladesh is densely populated, with 70 million people living in flood-prone areas. Flood events leave hundreds of thousands of people displaced yearly. A series of cyclones and storms over the years have led to mass destruction and loss of human life.

## Dataset Description

The Global Food Database was created by the Dartmouth Flood Observatory (DFO) and Cloud to Street with the intent of being utilized for global flood risk management and mitigation. This ImageCollection was funded by Google Earth Outreach, and includes 18 years (2000-2018) of satellite-based flood data available on Google Earth Engine. The dataset includes 913 flood events using 12,719 scenes that that were successfully mapped to include the extent and temporal distribution using Terra and Aqua MODIS sensors. Each of the events are represented as an image in the ImageCollection and can be filtered by date, country, and DFO “original ID”.

The dataset includes many image properties such as estimated fatalities, country code, main cause of flood, the centroid longitude and latitude, among others. An interesting image property that the visualizations will focus on is the severity of a flood event. It is broken down into 3 categories; 1 represents large floods that had significant damage and at least a 5-to-15-year interval since the last large flood, 1.5 is assigned to very large events with greater than 15 years but less than 100 year recurrence interval, and 2 denotes extreme events that have a recurrence interval of more than 100 years. 

The bands in the dataset include duration of flood along with a value of 0 or 1 to represent whether or not there was a flood event. Data on cloud coverage is displayed as the number of cloud-free observations in days between the start and end day of each event. In addition, cloud coverage has been assigned a percentage of clear view observations during a given flood event.

## Metadata Display and Basic Visualization

````python
## Map all floods to generate the satellite-observed historical flood plain.

#initialize map, centered around kansas ~ the geographic centroid of the USA
Map = geemap.Map(center=[-100, 40], zoom=4)

#using a satellite true color map
Map.setOptions('SATELLITE')

#set center of the map
Map.setCenter(-100, 40, 4)

#sum all of the flood events to get the entire historical flood plain
gfdFloodedSum = gfd.select('flooded').sum();

#create a palette for the raster data. Here, darker blue represents longer flood duration. 
durationPalette = ['C3EFFE', '1341E8', '051CB0', '001133', '#00020a']

#add the flood data layer to our previously initialized map
Map.addLayer(
  gfdFloodedSum.selfMask(),
  {'min': 0, 'max': 10, 'palette': durationPalette})

#plot the map
Map
````

````python
##Overlay permanent water to distinguish flood water from permanent water. 

#filters to just perm water. gte(1) means 
perm_water = gfd.select('jrc_perm_water').sum()

#adding the perm water later (a light cyan color) to the map
Map.addLayer(
  perm_water.selfMask(),
  {'min': 0, 'max': 1, 'palette': 'C3EFFE'},
  'JRC Permanent Water');

#plotting the map
Map
````

### Precipitation in Bangladesh

````python
# Point cooresponding to Bangladesh
pt = ee.Geometry.Point([90.3563, 23.6849])

# Define radius within which to grab data
scale = 100000
````

````python
gdat = ee.ImageCollection("IDAHO_EPSCOR/TERRACLIMATE")

# Select precipitation variable from ERA5 images
pr = gdat.select('pr')

# Store precipitation around Bangladesh as a list object
bd_pr_ts = pr.getRegion(pt, scale).getInfo()

# Turn precipitation information from GEE into a Pandas DataFrame
dataframe = pd.DataFrame(bd_pr_ts)
print(dataframe)
````

````python
headers = dataframe.loc[0]   
dataframe = pd.DataFrame(dataframe.values[1:], columns=headers)
````

````python
dataframe['datetime'] = pd.to_datetime(dataframe['time'], unit='ms')
dataframe
````

````python
# create a new figure, set size and resolution (dpi)
plt.figure(figsize=(10, 6), dpi=300) 

# add data to the plot
plt.plot(dataframe['datetime'],dataframe['pr'])   
plt.title('Bangladesh Rainfall', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Precipitation (mm)', fontsize=14)
plt.ylim(0, 1000)
plt.axvline(x = 10950, color = "red", label = "Start of Flood Data")
plt.legend()
````

## Use Case Examples

#### Comparing Individual Floods

One potential use case of this dataset is to compare the effects of individual flood events. 

Knowing how storms that are similar in nature vary in their destructiveness could prove useful in preparing certain cities/regions for future flooding events. For example, let’s say that Dhaka, the capital city of Bangladesh, experiences flooding due to monsoonal rain one year. As a result of the flooding, leaders in Bangladesh attempt to improve Dhaka’s drainage system. A few years later, the city experiences a monsoon of similar intensity as the previous one. This dataset could be utilized to see how the overall spread of the second flood compared.

On the flip side, comparing floods that are very different in nature could prove useful as well—looking at the cause of a flood and the spread of its impact could be helpful as a city, region, or country learns more about what types of extreme weather events are more likely to occur in our changing climate. Gaining an understanding of the different ways in which these extreme weather events affect flooding would be a step towards effective flood mitigation strategies. 

The code below demonstrates how one can isolate specific flood events and create a map to see where the flooding occurred. 


````python
bangladesh_aug14_id = 4178
bangladesh_aug14 = ee.Image(gfd.filterMetadata('id', 'equals', bangladesh_aug14_id).first())
Map = geemap.Map(center=[90, 24], zoom=6)
Map.setOptions('SATELLITE')
Map.setCenter(90, 24, 6)
Map.addLayer(bangladesh_aug14.select('flooded').selfMask(),
             {'min': 0, 'max': 1, 'palette': ['00FFFF', '0000FF']},
             'Bangladesh Flood - Inundation Extent')
Map
````

````python
bangladesh_jun03_id = 2248
bangladesh_jun03 = ee.Image(gfd.filterMetadata('id', 'equals', bangladesh_jun03_id).first())
Map = geemap.Map(center=[90, 24], zoom=6)
Map.setOptions('SATELLITE')
Map.setCenter(90, 24, 6)
Map.addLayer(bangladesh_jun03.select('flooded').selfMask(),
             {'min': 0, 'max': 1, 'palette': ['00FFFF', '0000FF']},
             'Bangladesh Flood - Inundation Extent')
Map
````

#### Isolating Time Periods to Track Flood Trends

Another important use case of this dataset is to compare flood events from various time periods to monitor long term flood patterns. 

Continuing our focus on Bangladesh, it would be helpful to observe where flooding is most likely to occur in the future so the government can respond accordingly; for example, bolstering infrastructure in certain regions could help prevent excessive damage from occurring while adjusting the location of farm plots (or adjusting the types of crops to be more flood resistant) could help prevent famine. 

Focusing on these longer-run trends could also reveal whether certain flood mitigation strategies employed by various regions are working. In an example above, we discussed how comparing individual flood events could be useful in determining the effect of new drainage system technology in Dhaka. We can take this example even further by examining an aggregation of flood events before and after the change in the drainage system occurred. 

The code below demonstrates how one can isolate specific time periods of interest to compare the flooding from those years. For our example, we compare flooding from the past two decades. 

````python
bangla_2000s = ee.ImageCollection("GLOBAL_FLOOD_DB/MODIS_EVENTS/V1").filterDate('2000-02-17', '2009-12-31')

# Initialize map, centered around kansas ~ the geographic centroid of the USA
Map = geemap.Map(center=[90, 24], zoom=6)

# Using a satellite true color map
Map.setOptions('SATELLITE')

# Set center of the map
Map.setCenter(90, 24, 6)

# Sum all of the flood events to get the entire historical flood plain
bangla_2000s_FloodedSum = bangla_2000s.select('flooded').sum();

# Create a palette for the raster data. Here, darker blue represents longer flood duration.
durationPalette = ['C3EFFE', '1341E8', '051CB0', '001133', '#00020A']

# Add the flood data layer to our previously initialized map
Map.addLayer(
  bangla_2000s_FloodedSum.selfMask(),
  {'min': 0, 'max': 10, 'palette': durationPalette})

## Overlay permanent water to distinguish flood water from permanent water.
# Filters to just perm water. gte(1) means
perm_water = bangla_2000s.select('jrc_perm_water').sum()

# Adding the perm water later (a light cyan color) to the map
Map.addLayer(
  perm_water.selfMask(),
  {'min': 0, 'max': 1, 'palette': 'C3EFFE'},
  'JRC Permanent Water');

# Plotting the map
Map
````

````python
bangla_2010s = ee.ImageCollection("GLOBAL_FLOOD_DB/MODIS_EVENTS/V1").filterDate('2010-01-01', '2018-12-10')

# Initialize map, centered around kansas ~ the geographic centroid of the USA
Map = geemap.Map(center=[90, 24], zoom=6)

# Using a satellite true color map
Map.setOptions('SATELLITE')

# Set center of the map
Map.setCenter(90, 24, 6)

# Sum all of the flood events to get the entire historical flood plain
bangla_2010s_FloodedSum = bangla_2010s.select('flooded').sum();

# Create a palette for the raster data. Here, darker blue represents longer flood duration.
durationPalette = ['C3EFFE', '1341E8', '051CB0', '001133', '#00020A']

# Add the flood data layer to our previously initialized map
Map.addLayer(
  bangla_2010s_FloodedSum.selfMask(),
  {'min': 0, 'max': 10, 'palette': durationPalette})

## Overlay permanent water to distinguish flood water from permanent water.
# Filters to just perm water. gte(1) means
perm_water = bangla_2010s.select('jrc_perm_water').sum()
 
# Adding the perm water later (a light cyan color) to the map
Map.addLayer(
  perm_water.selfMask(),
  {'min': 0, 'max': 1, 'palette': 'C3EFFE'},
  'JRC Permanent Water');

# Plotting the map
Map
````
